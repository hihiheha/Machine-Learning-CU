```python
# Importing necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv('/Users/eva/Downloads/archive-2/club_churn_test.csv')

# Confirm column types and convert if necessary
print("Data Types Before Conversion:")
print(data.dtypes)

data['MEMBERSHIP_TERM_YEARS'] = pd.to_numeric(data['MEMBERSHIP_TERM_YEARS'], errors='coerce')
data['ANNUAL_FEES'] = pd.to_numeric(data['ANNUAL_FEES'], errors='coerce')

# Drop rows with invalid data in critical columns
data.dropna(subset=['MEMBERSHIP_TERM_YEARS', 'ANNUAL_FEES'], inplace=True)

# Exploratory Data Analysis (EDA)
print("\nDataset Overview After Cleaning:")
print(data.head())

# Distribution of Membership Term Years
plt.figure(figsize=(8, 5))
plt.hist(data['MEMBERSHIP_TERM_YEARS'], bins=20, color='blue', edgecolor='black')
plt.title('Distribution of Membership Term Years')
plt.xlabel('Membership Term Years')
plt.ylabel('Frequency')
plt.show()

# Distribution of Annual Fees
plt.figure(figsize=(8, 5))
plt.hist(data['ANNUAL_FEES'], bins=20, color='green', edgecolor='black')
plt.title('Distribution of Annual Fees')
plt.xlabel('Annual Fees')
plt.ylabel('Frequency')
plt.show()

# Gender Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='MEMBER_GENDER', data=data, palette='coolwarm', hue='MEMBER_GENDER', dodge=False)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend([], [], frameon=False)  
plt.show()

# Data Preprocessing and Cleaning
data.drop(columns=['Unnamed: 0', 'MEMBERSHIP_NUMBER', 'START_DATE', 'END_DATE', 'AGENT_CODE'], inplace=True)

# Handle missing values
categorical_cols = ['MEMBER_MARITAL_STATUS', 'MEMBER_GENDER', 'MEMBERSHIP_PACKAGE', 'PAYMENT_MODE']
numerical_cols = ['MEMBER_ANNUAL_INCOME', 'MEMBER_OCCUPATION_CD']

for col in categorical_cols:
    data[col] = data[col].fillna(data[col].mode()[0])
for col in numerical_cols:
    data[col] = data[col].fillna(data[col].mean())

# One-hot encode categorical columns
data_encoded = pd.get_dummies(data, drop_first=True)

# Features and Target
X = data_encoded.drop(columns=['ANNUAL_FEES'])
y = (data_encoded['ANNUAL_FEES'] > data_encoded['ANNUAL_FEES'].median()).astype(int)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Model 1: Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
dt_predictions = dt_model.predict(X_test)

# Model 2: K-Nearest Neighbors
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)
knn_predictions = knn_model.predict(X_test)

# Results
print("\nDecision Tree Accuracy:", accuracy_score(y_test, dt_predictions))
print("KNN Accuracy:", accuracy_score(y_test, knn_predictions))

# Classification Reports
print("\nDecision Tree Report:")
print(classification_report(y_test, dt_predictions))
print("\nKNN Report:")
print(classification_report(y_test, knn_predictions))

# Confusion Matrices Visualization
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
sns.heatmap(confusion_matrix(y_test, dt_predictions), annot=True, fmt="d", cmap="Blues", ax=axes[0])
axes[0].set_title("Decision Tree Confusion Matrix")
sns.heatmap(confusion_matrix(y_test, knn_predictions), annot=True, fmt="d", cmap="Greens", ax=axes[1])
axes[1].set_title("KNN Confusion Matrix")
plt.show()

# Decision Tree Visualization
plt.figure(figsize=(20, 10))
plot_tree(dt_model, feature_names=X.columns, class_names=['Low Fees', 'High Fees'], filled=True, fontsize=10)
plt.title('Decision Tree Visualization')
plt.show()
```
